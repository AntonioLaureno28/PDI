```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
# Carregando a imagem com o OpenCV
imagem  = cv2.imread('IO.jpg')
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

image.shape # Dimens√µes da imagem
```

```python
def redimensionar(image, fator):
    altura, largura = image.shape

    Nova_altura = int(altura*fator)
    Nova_largura = int(largura*fator)

    nova_imagem = np.zeros((Nova_altura, Nova_largura))

    for y_1 in range(Nova_altura):
        for x_1 in range(Nova_largura):
            x_0 = int(round(x_1/fator))
            y_0 = int(round(y_1/fator))

            y_0 = min(y_0, altura - 1) 
            x_0 = min(x_0, largura - 1)

            nova_imagem[y_1, x_1] = image[y_0, x_0]

    return nova_imagem
    
```

```python
def reduzir_imagem(image, fator):
    fator_reducao = 1 / fator
    nova_imagem = redimensionar(image, fator_reducao)
    return nova_imagem
```

```python
def ampliar_imagem(image, fator):
    nova_imagem = redimensionar(image, fator)
    return nova_imagem
```

```python
imagem_reduzida = reduzir_imagem(image, 10)

plt.figure(figsize=(8, 6))
plt.imshow(imagem_reduzida, cmap='gray')
plt.title('Imagem Reduzida')

imagem_ampliada = ampliar_imagem(imagem_reduzida, 10)

plt.figure(figsize=(8, 6))
plt.imshow(imagem_ampliada, cmap='gray')
plt.title('imagem Ampliada')
```

```python
fig, eixos = plt.subplots(1, 3, figsize=(15,10))

eixos[0].imshow(image, cmap='gray')
eixos[0].set_title('Imagem Original')

eixos[1].imshow(imagem_reduzida, cmap='gray')
eixos[1].set_title(f'Imagem Reduzida em um fator de 10')

eixos[2].imshow(imagem_ampliada, cmap='gray')
eixos[2].set_title(f'Imagem Reampliada em um fator de 10')
```
