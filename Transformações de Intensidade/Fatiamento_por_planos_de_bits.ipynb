```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

## Fatiamento por planos de bits na imagem 3.14

```python
# Carregando a imagem com o OpenCV
imagem  = cv2.imread('images/dollar.jpg')
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

image.shape
```

```python
def Fatiamento_por_plano_de_bit(image):
    
    num_bits = 8
    
    planos_de_bits = [] 
    
    for k in range(num_bits):
        
        bit_plano_k = (image >> k) & 1 # Realizando o bitwise e aplicando o AND 
        
        plano_visualizacao = bit_plano_k.astype(np.uint8) * 255 # Convertendo o array binário para uma imagem vizualizável
        
        planos_de_bits.append(plano_visualizacao)
        
    return planos_de_bits
```

```python
planos = Fatiamento_por_plano_de_bit(imagem)

plt.figure(figsize=(12, 6))
plt.suptitle('Planos de Bit', fontsize=16)

plt.subplot(3, 3, 1) 
plt.imshow(imagem, cmap='gray') 
plt.axis('off')

plt.subplot(3, 3, 2) 
plt.imshow(planos[0], cmap='gray') 
plt.axis('off')

plt.subplot(3, 3, 3) 
plt.imshow(planos[1], cmap='gray') 
plt.axis('off')

plt.subplot(3, 3, 4) 
plt.imshow(planos[2], cmap='gray') 
plt.axis('off')

plt.subplot(3, 3, 5) 
plt.imshow(planos[3], cmap='gray') 
plt.axis('off')

plt.subplot(3, 3, 6) 
plt.imshow(planos[4], cmap='gray') 
plt.axis('off')

plt.subplot(3, 3, 7) 
plt.imshow(planos[5], cmap='gray') 
plt.axis('off')

plt.subplot(3, 3, 8) 
plt.imshow(planos[6], cmap='gray') 
plt.axis('off')

plt.subplot(3, 3, 9) 
plt.imshow(planos[7], cmap='gray') 
plt.axis('off')


plt.tight_layout()
plt.show()
```
