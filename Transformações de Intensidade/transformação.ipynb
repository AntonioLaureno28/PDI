```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

## Transformações Gama aplicadas nas imagens 3.8 e 3.9 do livro

### Aplicando na Imagem 3.8

```python
# Carregando as imagens com o OpenCV
imagem  = cv2.imread('images\Spine.png')
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

imagem1 = cv2.imread('images\Aerial.jpg')
image1 = cv2.cvtColor(imagem1, cv2.COLOR_BGR2GRAY)

image.shape, image1.shape # Dimensões das imagens
```

```python
def transformacao_gama(image, gama, c=1.0):

    r_normalized = np.arange(0, 256) / 255.0 # Entrada Normalizada com um array de 256 valores

    s_normalized = c * (r_normalized**gama)

    s = s_normalized * 255.0 # Convertendo os resultados de volta 

    tabela_gama = np.clip(s, 0, 255).astype('uint8')

    imagem_transformada = cv2.LUT(image, tabela_gama) # Função do Opencv que mapea todos os valores de pixels de uma imagem para novos valores

    return imagem_transformada 
```

```python
gama1 = 0.6
img_gama1 = transformacao_gama(image, gama1)

gama2 = 0.4
img_gama2 = transformacao_gama(image, gama2)

gama3 = 0.3
img_gama3 = transformacao_gama(image, gama3)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_gama1, cmap='gray')
plt.title(f'Gama = {gama1}')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_gama2, cmap='gray')
plt.title(f'Gama = {gama2}')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_gama3, cmap='gray')
plt.title(f'Gama = {gama3}')
plt.axis('off')

plt.show()
```

### Aplicando na imagem 3.9

```python
gama_1 = 3.0
img_gama_corrigida_1 = transformacao_gama(image1, gama_1)

gama_2 = 4.0
img_gama_corrigida_2 = transformacao_gama(image1, gama_2)

gama_3 = 5.0
img_gama_corrigida_3 = transformacao_gama(image1, gama_3)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(image1, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_gama_corrigida_1, cmap='gray')
plt.title(f'Gama = {gama_1}')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_gama_corrigida_2, cmap='gray')
plt.title(f'Gama = {gama_2}')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_gama_corrigida_3, cmap='gray')
plt.title(f'Gama = {gama_3}')
plt.axis('off')

plt.show()
```
