```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
# Carregando a imagem com o OpenCV
imagem  = cv2.imread('images/Circuito.jpg')
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

image.shape
```

## Aplicando os filtros

```python
imagem_filtrada_por_media_movel = cv2.blur(image, (3, 3)) # Filtro de média móvel (3x3)

imagem_filtrada_por_mediana = cv2.medianBlur(image, 3) # Filtro de mediana (3x3)
```

## Plotando os resultados

```python
plt.figure(figsize=(15,10))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagem original')


plt.subplot(1, 3, 2)
plt.imshow(imagem_filtrada_por_media_movel, cmap='gray')
plt.title('Imagem com filtro de média')


plt.subplot(1, 3, 3)
plt.imshow(imagem_filtrada_por_mediana, cmap='gray')
plt.title('Imagem com filtro de mediana')
```
