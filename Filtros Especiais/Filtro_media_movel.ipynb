```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
# Carregando a imagem com o OpenCV
imagem  = cv2.imread('images/hubble.jpg')
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

image.shape
```

## Aplicando o Filtro de m√©dia

```python
imagem_corrigida = cv2.blur(image, (15, 15))
```

```python
# Plotando o resultado

plt.figure(figsize=(10,8))
plt.imshow(imagem_corrigida, 'gray')
```

## Aplicando o a limiariza√ß√£o

```python
t = 65
binary_mask = imagem_corrigida > t

plt.figure(figsize=(10,8))
plt.imshow(binary_mask, cmap='gray')
```

## Plotando as imagens

```python
plt.figure(figsize=(15,15))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagem original')

plt.subplot(1, 3, 2)
plt.imshow(imagem_corrigida, cmap='gray')
plt.title('Imagem filtrada com filtro de m√©dia (15x15)')


plt.subplot(1, 3, 3)
plt.imshow(binary_mask, cmap='gray')
plt.title('Imagem filtrada com limiariza√ß√£o')
```
